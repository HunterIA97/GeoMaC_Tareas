```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#funci칩n 1
f1=lambda x:1-np.exp(x)+(np.exp(1)-1)*np.sin(np.pi*x/2)
x=np.linspace(-2,1,100)
plt.plot(x,f1(x))
plt.grid()
plt.show()
```

```python
#funci칩n 2
f2=lambda x:(x-1)*np.tan(x)+x*np.sin(np.pi*x)
x=np.linspace(-2,-1,100)
plt.plot(x,f2(x))
plt.grid()
plt.show()
```

```python
#funci칩n 3
f3=lambda x:x*np.sin(np.pi*x)-(x-2)*np.log(x)
x=np.linspace(1,3,100)
plt.plot(x,f3(x))
plt.grid()
plt.show()
```

```python
#funci칩n 4
f4=lambda x:(x-2)*np.sin(x)*np.log(x+2)
x=np.linspace(-1,3,100)
plt.plot(x,f4(x))
plt.grid()
plt.show()
```
